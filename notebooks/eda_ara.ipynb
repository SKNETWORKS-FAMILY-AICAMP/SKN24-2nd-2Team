```python
# 이상치 확인
# age
print(df['age'].value_counts().tail(10))

# height
print(df['height'].value_counts().head(10))

# income
print(df['income'].value_counts().head(10))
```

```python
# 이상값 확인 및 처리

print("age 이상값 확인")
print(f"18살 미만: {(df['age'] < 18).sum()}명")
print(f"60살 초과: {(df['age'] > 60).sum()}명")

# 18~60살로 필터링
df = df[(df['age'] >= 18) & (df['age'] <= 60)]
print(f"처리 후 데이터 수: {len(df)}명")
```

```python
# height 이상값 확인 & 처리
print("\nheight 이상값 확인")
print(f"최소값: {df['height'].min()} 인치")
print(f"최대값: {df['height'].max()} 인치")
print(f"59인치 미만: {(df['height'] < 59).sum()}명")  # 150cm 미만
print(f"79인치 초과: {(df['height'] > 79).sum()}명")  # 200cm 초과

# 59~79인치로 필터링 (결측치 유지)
df = df[(df['height'] >= 59) & (df['height'] <= 79) | df['height'].isna()]
print(f"처리 후 데이터 수: {len(df)}명")

# 범위 재확인
print(f"\nheight 범위: {df['height'].min()} ~ {df['height'].max()} 인치")
print(f"height 범위: {df['height'].min() * 2.54:.1f} ~ {df['height'].max() * 2.54:.1f} cm")
```

```python

# income 이상값 처리

print("\nincome -1 개수:", (df['income'] == -1).sum())

# -1 을 NaN으로 변환
df['income'] = df['income'].replace(-1, np.nan)
print("NaN 변환 후 결측치 수:", df['income'].isna().sum())
```

```python

# 이상값 제거 처리 후 수치형 분포 그래프 

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.histplot(df['age'], kde=True, ax=axes[0], color='steelblue')
axes[0].set_title('나이 분포 (이상값 제거 후)')

sns.histplot(df['height'].dropna(), kde=True, ax=axes[1], color='steelblue')
axes[1].set_title('키 분포 (이상값 제거 후)')

sns.histplot(df['income'].dropna(), kde=True, ax=axes[2], color='steelblue')
axes[2].set_title('수입 분포 (결측치 제거 후)')

plt.tight_layout()
plt.show()
```

```python
print("최종 데이터 shape:", df.shape)
print("\nage 범위:", df['age'].min(), "~", df['age'].max())
print("height 범위:", df['height'].min(), "~", df['height'].max())
print("income 결측치 수:", df['income'].isna().sum())
```

```python
# 결측치 확인
dt = pd.DataFrame({
    '결측치 수': df.isnull().sum(),
    '결측치 비율(%)': (df.isnull().sum() / len(df) * 100).round(2)
})

print(dt[dt['결측치 수'] >0].sort_values('결측치 비율(%)', ascending=False))
```

```python
# 결측치 히트맵
plt.figure(figsize=(14, 6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
plt.title('결측치 히트맵')
plt.show()
```

```python
# 결측치 비율 막대그래프
dt = (df.isnull().sum() / len(df) * 100).sort_values(ascending=False)
dt = dt[dt > 0]  # 결측치 있는 컬럼만

plt.figure(figsize=(12, 6))
sns.barplot(x=dt.values, y=dt.index, color='skyblue')
plt.title('컬럼별 결측치 비율(%)')
plt.xlabel('결측치 비율(%)')
for i, v in enumerate(dt.values):
    plt.text(v + 0.5, i, f'{v:.1f}%', va='center')
plt.tight_layout()
plt.show()
```

```python
missing_df = pd.DataFrame({
    '결측치 수': df.isnull().sum(),
    '결측치 비율(%)': (df.isnull().sum() / len(df) * 100).round(2)
})
missing_df = missing_df[missing_df['결측치 수'] > 0].sort_values('결측치 비율(%)', ascending=False)
print(missing_df)
```

```python
# 수치형 컬럼 분포  
# income -1 은 결측치이므로 제외
df_income = df[df['income'] != -1]

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.histplot(df['age'], kde=True, ax=axes[0])
axes[0].set_title('나이 분포')

sns.histplot(df['height'].dropna(), kde=True, ax=axes[1])
axes[1].set_title('키 분포')

sns.histplot(df_income['income'], kde=True, ax=axes[2])
axes[2].set_title('수입 분포')

plt.tight_layout()
plt.show()
```

```python
# 범주형 컬럼 분포
cat_cols = ['sex', 'orientation', 'status', 'body_type', 'education', 'drinks']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(cat_cols):
    df[col].value_counts().plot(kind='bar', ax=axes[i], color='steelblue')
    axes[i].set_title(f'{col} 분포')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
