```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/okcupid_preprocessed_revised.csv')
```

```python
df.info()
```

```python
df = df.drop(columns="Unnamed: 0")
```

```python
df.info()
```

```python
# X 데이터 정리
drop_cols = [
    "ethnicity_black", "ethnicity_hispanic / latin", "ethnicity_mixed", "ethnicity_other","ethnicity_white",
    "dogs_encoding", "cats_encoding", "location_group_north_bay", "location_group_outside", "location_group_sf",
    "location_group_south_bay", "sign_1", "last_online", "job"
]

drop_list = []

for col in drop_cols:
    if col in df.columns:
        drop_list.append(col)

df = df.drop(columns=drop_list)
```

```python
df.info()
```

```python
df = df.drop(columns="stability_index")
```

```python
sns.scatterplot(
    x="total_essay_len",
    y="Response rate",
    hue="is_churned",
    data=df,
    alpha=0.3
)
plt.show()
```

```python
sns.scatterplot(
    x="essay_answered_count",
    y="total_essay_len",
    hue="is_churned",
    data=df,
    alpha=0.3
)
plt.show()
```

```python
sns.jointplot(
    x="total_essay_len",
    y="Response rate",
    hue="is_churned",
    data=df,
    alpha=0.3
)
```
